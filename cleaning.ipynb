### Importing the required libraries

```python
import pandas as pd
```

### Load the raw CSV into a Pandas Dataframe 

```python
df = pd.read_csv("universities.csv", encoding="utf-8")
df.head()
```

```python
df.info()
```

### Clean the data

```python
df["Founded"].values
```

```python
#extract the first 4 digits in '1538 (reopened 1914)'

df["Founded"] = df["Founded"].str.extract(r"(\d{4})")
```

```python
df["Founded"].values
```

```python
#coverts column to interger type

df["Founded"] = df["Founded"].astype(int)
```

```python
df.info()
```

```python
df["Enrollment"].values
```

```python
df["Enrollment"] = df["Enrollment"].str.replace(r"[^\d]", "", regex=True).astype(int).values
```

```python
df["Enrollment"].values
```

```python
# Seperating the location column to city and country

df["Location"].values
```

```python
df[["location_city", "location_country"]] = df["Location"].str.extract(r"^(.*?),?\s*([^,]+)$")
df
```

```python
df["location_city"] = df["location_city"].replace('', pd.NA).fillna('N/A')
```

```python
df.describe()
```

```python
df.info()
```

```python
# Manually setting the updated year to 2025.

df["updated_year"] = "2025"
print(df.columns)
```
